# Project 3 - AI Recommendation Logic

This notebook builds a simple content-based recommendation system. It takes user interests, compares them with item attributes, calculates similarity scores, and displays the best matching recommendations.

## Goal

- Take user input as choices or interests
- Match preferences using similarity logic
- Display recommended items

## Recommendation Logic

```python
def normalize_preferences(raw_preferences):
    """Convert comma-separated user input into a clean set of lowercase preferences."""
    return {
        preference.strip().lower()
        for preference in raw_preferences.split(",")
        if preference.strip()
    }


def similarity_score(preferences, item):
    """Score an item using tag overlap plus small category and difficulty bonuses."""
    if not preferences:
        return 0.0

    overlap = preferences.intersection(item["tags"])
    union = preferences.union(item["tags"])
    score = len(overlap) / len(union)

    if item["category"].lower() in preferences:
        score += 0.15

    if item["difficulty"].lower() in preferences:
        score += 0.10

    return round(score, 4)


def recommend_items(raw_preferences, limit=5):
    preferences = normalize_preferences(raw_preferences)

    results = []
    for item in catalog:
        score = similarity_score(preferences, item)
        if score > 0:
            results.append({**item, "similarity_score": score})

    results.sort(key=lambda item: item["similarity_score"], reverse=True)
    return pd.DataFrame(results[:limit])
```

## Test With User Preferences

```python
user_preferences = "ai, python, beginner, classification"

recommendations = recommend_items(user_preferences)
recommendations[["title", "category", "difficulty", "similarity_score", "description"]]
```
