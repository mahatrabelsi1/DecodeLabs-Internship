# Project 4 - Image Recognition Basic

This notebook implements a basic image recognition task using scikit-learn. The model learns to recognize handwritten digit images and displays predictions clearly on sample input.

## Project Requirements

- Use a pre-trained model or simple AI library
- Perform recognition on sample input
- Display the output clearly

```python
import os

os.environ.setdefault("LOKY_MAX_CPU_COUNT", "1")

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_digits
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

## Load Sample Image Dataset

```python
digits = load_digits()

X = digits.data
y = digits.target
images = digits.images

print("Dataset shape:", X.shape)
print("Image shape:", images[0].shape)
print("Target classes:", digits.target_names)
```

## View Sample Images

```python
fig, axes = plt.subplots(1, 6, figsize=(9, 2))

for axis, image, label in zip(axes, images[:6], y[:6]):
    axis.imshow(image, cmap="gray")
    axis.set_title(f"Digit: {label}")
    axis.axis("off")

plt.tight_layout()
plt.show()
```

## Train Recognition Model

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y,
)

model = Pipeline(
    steps=[
        ("scaler", StandardScaler()),
        ("classifier", KNeighborsClassifier(n_neighbors=5)),
    ]
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Model accuracy: {accuracy:.2%}")
```

## Recognize a Sample Image

```python
sample_index = 25
sample_image = images[sample_index]
sample_data = X[sample_index]

predicted_digit = model.predict([sample_data])[0]
probabilities = model.predict_proba([sample_data])[0]
confidence = probabilities[predicted_digit]

plt.figure(figsize=(3, 3))
plt.imshow(sample_image, cmap="gray")
plt.title(f"Predicted: {predicted_digit} | Actual: {y[sample_index]}")
plt.axis("off")
plt.show()

print("Recognition Output")
print("Sample index:", sample_index)
print("Actual digit:", y[sample_index])
print("Predicted digit:", predicted_digit)
print(f"Confidence: {confidence:.2%}")
```

## Evaluate Recognition Results

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 5))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=digits.target_names,
    yticklabels=digits.target_names,
)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Digit")
plt.ylabel("Actual Digit")
plt.tight_layout()
plt.show()

print(classification_report(y_test, y_pred))
```

## Conclusion

The model recognizes handwritten digit images by learning patterns in pixel values. The sample prediction and evaluation metrics show how image recognition outputs can be interpreted clearly.
